```python
import re
def timeConvert(text):
    text=text.replace(chr(0x202F)," ")
    mytime=text.split(" ")
    if mytime[2][-2]=='a':
        if int(mytime[1].split(":")[0])==12:
            return "".join(mytime[0]+f"00:{mytime[1].split(':')[1]}")
        return "".join(mytime[0:2])
    else:
        gethour=int(re.findall('^\d{1,2}',mytime[1])[0])
        if gethour==12:
            return "".join(mytime[0]+f"12:{mytime[1].split(':')[1]}")
        replace24=gethour+12
        replace24=str(replace24)
        return "".join(mytime[0]+f"{replace24}:{mytime[1].split(':')[1]}")
        
```

```python
df['message_date']=df['message_date'].apply(timeConvert)
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def transform_text(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    
    y=[]
    for item in text:
        if item.isalnum():
            y.append(item)
    text=y.copy()
    y.clear()
    for item in text:
        if item not in stopwords.words('english') and item not in string.punctuation:
            y.append(item)
    text=y.copy()
    y.clear()
    for item in text:
        y.append(ps.stem(item))
    return " ".join(y)
```

```python
df['transformed']=df['user_messages'].apply(transform_text)
```
